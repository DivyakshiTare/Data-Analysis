```python
import pandas as pd
df = pd.read_csv("/content/SocialMediaUsers.csv")
```

```python
print(df.info())
```

```python
print(df.head())
```

```python
import matplotlib.pyplot as plt

df['Age'] = (pd.to_datetime('now') - pd.to_datetime(df['DOB'])).astype('<m8[Y]')

plt.hist(df['Age'], bins=15, edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distribution of Social Media Users')
plt.show()
```

```python
df['Gender'].value_counts().plot(kind='bar', color=['lightblue', 'pink'])
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution of Social Media Users')
plt.show()
```

```python
top_countries = df['Country'].value_counts().head(10)
top_countries.plot(kind='bar', color='skyblue')
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Top 10 Countries of Social Media Users')
plt.show()
```

```python
all_interests = [Interest for Interests in df['Interests'].dropna() for Interest in Interests.split(',')]
interest_counts = pd.Series(all_interests).value_counts()


interest_counts.head(10).plot(kind='bar', color='orange')
plt.xlabel('Interest')
plt.ylabel('Count')
plt.title('Top 10 Interests of Social Media Users')
plt.show()
```
